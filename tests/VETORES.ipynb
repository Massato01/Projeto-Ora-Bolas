```python
from math import *

def velocidade_media(S0i=1, S1i=2, S0j=1, S1j=2,
                     Ti=1, Tf=2):
    var_t = (Tf - Ti)
    var_s_i = (S1i - S0i) / var_t
    var_s_j = (S1j - S0j) / var_t
    return f'{var_s_i:.2f}i   {var_s_j:.2f}j'


def F_ax(F=1, ang=1):
    return F * cos(ang)


def F_ay(F=1, ang=1):
    return F * sin(ang)


def A_vetorial(Axi=0, Ayj=0):
    return Axi + Ayj


def B_vetorial(Bxi=0, Byj=0):
    return Bxi + Byj


def vetor_soma(Ai=0, Aj=0, Bi=0, Bj=0):
    return f'{Ai + Bi}i  {Aj + Bj}j'


def Rx_modulo(Ax=0, Bx=0):
    return Ax + Bx


def Ry_modulo(Ay=0, By=0):
    return Ay + By


def modulo(Rx=0, Ry=0):
    return sqrt((Rx ** 2) + (Ry ** 2))


def direcao(Rx=1, Ry=1):
    arcotan = atan(Ry / Rx)
    return arcotan * 180 / pi


# Main
print('--- Vel Md ---')
print(f'VELOCIDADE MEDIA: {velocidade_media(15, 45, -15, 45, 100, 1000)}')
print('-------------')
print('--- FORCA ---')
print(f'Fax: {F_ax()}')
print(f'Fay: {F_ay()}')
print('-------------')
print('--- VETOR ---')
print(f'VETOR A: {A_vetorial()}')
print(f'VETOR B: {B_vetorial()}')
print(f'VETOR SOMA: {vetor_soma(4, -3, -1, 1)}')
print('-------------')
print('-- MODULO --')
print(f'RX: {Rx_modulo()}')
print(f'RY: {Ry_modulo()}')
print(f'MODULO: {modulo(-8.60, 5.20):.3f}')
print('-------------')
print('-- DIRECAO --')
print(f'DIRECAO: {direcao(-8.60, 5.20):.1f}')
print('-------------')
```
